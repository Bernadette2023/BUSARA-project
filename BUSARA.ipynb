c) Data Cleaning

As seen from the description of the dataset , there is alot of cleaning to be done .The first thing to clean is the Outliers. This is clear on the Maximum especially for age ,height and Weight. It shows that :


Maximum Age is 544
Maximum Weight is 79174
Maximum Height is 1.56


As much as its possible for Age and Weight to be of that vaIue ,it might not be plausible .Inorder to have clear and consice data , this has to be rectrified . I went ahead and Filtered the data to a more reasonable number .

```python

#Assuming this are reasonable numbers
age_range = (18, 100)  
height_range = (100, 220)  
weight_range = (20, 300) 

# Filtering out data points outside the reasonable ranges
filtered_data = df[(df['Age'] >= age_range[0]) & (df['Age'] <= age_range[1]) &
                     (df['Height'] >= height_range[0]) & (df['Height'] <= height_range[1]) &
                     (df['Weight_'] >= weight_range[0]) & (df['Weight_'] <= weight_range[1])]

# Print the number of data points before and after filtering
print("Number of data points before filtering:", len(df))
print("Number of data points after filtering:", len(filtered_data))

# Show the filtered data
filtered_data.head()
```
